```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("ecommerce_sales_raw.csv")
print("First 5 rows:")
print(df.head())
print("\nColumn names:")
print(df.columns)
print("\nMissing values per column:")
print(df.isnull().sum())
df = df.drop_duplicates()
print("\nShape after removing duplicates:", df.shape)
df["category"] = df["category"].str.strip().str.title()
df["unit_price"] = df["unit_price"].fillna(df["unit_price"].median())
df["customer_rating"] = df["customer_rating"].fillna(df["customer_rating"].median())
df["customer_age"] = df["customer_age"].fillna(df["customer_age"].median())
df = df.dropna(subset=["region"])
```

```python
# Bar Chart — Category Distribution
df["category"].value_counts().plot(kind="bar", color="steelblue")
plt.title("Category Distribution")
plt.xlabel("Category")
plt.ylabel("Count")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
# Pie Chart — Payment Method Share
df["payment_method"].value_counts().plot(kind="pie", autopct="%1.0f%%")
plt.title("Payment Method Share")
plt.ylabel("")
plt.tight_layout()
plt.show()
```

```python
# Histogram — Order Value Distribution
df["total_price"].plot(kind="hist", bins=30, color="mediumpurple")
plt.title("Distribution of Order Value")
plt.xlabel("Order Total ($)")
plt.tight_layout()
plt.show()
```
